```python
import pandas as pd
```

```python
res2 = pd.read_html('https://www.ulsan.go.kr/u/health/contents.ulsan?mId=001002003000000000', flavor = 'bs4', encoding='utf-8')
df = res2[1]
df
```

```python
df['도시'] = df['환자'].str.split('#').str[0]
df['확진자번호'] = df['환자'].str.split('#').str[1]
df
```

```python
del_indexs = df[ df['도시'] != '울산' ].index
df.drop(del_indexs , inplace=True)
df
```

```python
del_indexs
```

```python
df.dropna(inplace=True)
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df['환자'] = df['환자'].str.split('#').str[1]
df
```

```python
df['거주지'].value_counts()
```

```python
df['거주지'] = df['거주지'].str.split(' ').str[0]
df
```

```python
df['거주지'] = df['거주지'].str.split('(').str[0]
df
```

```python
df['격리시설/퇴원일'] = df['격리시설/퇴원일'].str.split('(').str[0]
df
```

```python
from pandas import DataFrame as dff

df2=df['거주지'].value_counts()

df2=dff(data={'거주지':df2.index, '확진자':df2.values})

df2
```

```python
is_today=df['확진일']=='1/14'

today=df[is_today]

today
```

```python
df3=today['거주지'].value_counts()

df3=dff(data={'거주지':df3.index, '추가 확진자':df3.values})

df3
```

```python
df4=pd.merge(df2, df3, on='거주지', how='left')
df4=df4.fillna(0)

df4
```

```python
df4.to_csv('covid_uls_0114.csv')
```
